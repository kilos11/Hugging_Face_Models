```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load the pre-trained sentence transformer model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Example user preferences
user_preferences = [
    "I like action movies with intense fight scenes and car chases.",
    "I enjoy romantic comedies with witty dialogues and heartwarming stories."
]

# Example movie descriptions
movie_descriptions = [
    "An ex-hitman comes out of retirement to track down the gangsters that took everything from him.",
    "A young woman disregards the advice of others so she can marry a handsome aristocrat she just met in Europe.",
    "A struggling writer finds inspiration and love after moving to a charming bookstore in a small town."
]

# Compute embeddings for user preferences and movie descriptions
user_embeddings = model.encode(user_preferences)
movie_embeddings = model.encode(movie_descriptions)

# Compute cosine similarity between user preferences and movie descriptions
similarities = np.dot(user_embeddings, movie_embeddings.T)

# Recommend the most similar movies for each user preference
for i, user_embedding in enumerate(user_embeddings):
    movie_scores = similarities[i]
    sorted_indices = np.argsort(-movie_scores)
    print(f"Recommendations for preference '{user_preferences[i]}':")
    for j in sorted_indices[:2]:
        print(f"- {movie_descriptions[j]}")
    print()
```
